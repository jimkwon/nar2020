## RPM data for K562 (+293E)

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
```

```python
import os
import glob

sWorkingDir = '/casa/jimkwon/PROJECT/ERH/data/200110_hiseq/analysis/annotated/'

lSamples = ['A1siGFP','A2siNC','A3siERH3',
            'B1siGFP','B2siNC','B3siERH3','B4siERH5',
            'BK16siGFP','BK19siERH5']
fPseudoCount = 1

lCountDataFrames = []
for sSample in lSamples:
    print(sSample)
    sMirnaCountFile = os.path.join(sWorkingDir, sSample + '.mirna.count.txt')
    
    dfCount = pd.read_table(sMirnaCountFile, names=[sSample, 'name'], sep=' ', skipinitialspace=True)
    dfCount['name'] = dfCount['name'].apply(lambda x: x.split('Name=')[1])
        
    dfCount = dfCount.set_index('name')
    lCountDataFrames.append(dfCount)

dfMerged = lCountDataFrames[0].join(lCountDataFrames[1:], how='outer', sort=True)
dfMerged = dfMerged.fillna(0) + fPseudoCount

dfMerged
```

```python
lSum = dfMerged.sum(axis=0)
lSum
```

```python
nScale = 1000000
lNames = ['A1siGFP','A2siNC','A3siERH3',
            'B1siGFP','B2siNC','B3siERH3','B4siERH5',
            'BK16siGFP','BK19siERH5']

for i, sName in enumerate(lNames):
    print(i, sName)
    
    fSum = lSum[i]
    
    dfMerged[sName + '_norm'] = dfMerged[sName].apply(lambda row: row / fSum * nScale)
    
dfMerged
```

```python
nSamples = len(lNames) # from previous column
dfNormCount = dfMerged.iloc[:, nSamples:]
dfNormCount
```

```python
dfLogNormCount = dfNormCount.apply(lambda x: np.log2(x), axis=1)
dfLogNormCount
```

## From above analysis, now I have these data frames.

Normalization: (Count + 1) / sum of reads * 1,000,000

dfLogNormCount (log2 normalized counts)

```python
from scipy.stats import spearmanr as spearmanr
from scipy.stats import pearsonr as pearsonr
import numpy as np 
```

## figure for paper

```python

lCombi = [ ['B1siGFP', 'B2siNC'],           
           ['B3siERH3', 'B4siERH5'],
           ['B1siGFP', 'B4siERH5']]

sMirInterest = \
"""
hsa-mir-181b-1
hsa-mir-15a
"""

lMirInterest = []
for sMir in sMirInterest.split():
    sMir = sMir.strip()
    lMirInterest.append(sMir)
dfLogNormCount_only_interest = dfLogNormCount.loc[lMirInterest,:]
    
fig = plt.figure(figsize=(9,9))

for n, (i, j) in enumerate(lCombi):    
    
    ax = fig.add_subplot(2,2,n+1)
    
    i_count = dfLogNormCount[i+'_norm'].tolist()
    j_count = dfLogNormCount[j+'_norm'].tolist()
    
    fPearsonRho, fPearsonP = pearsonr(i_count, j_count)
    
    i_inter = dfLogNormCount_only_interest[i+'_norm'].tolist()
    j_inter = dfLogNormCount_only_interest[j+'_norm'].tolist()
    name_inter = dfLogNormCount_only_interest.index.tolist()
    
    ax.plot([2,18], [2,18], color='gray', linestyle='--', linewidth=1)
    ax.scatter(i_count, j_count, color='black', alpha=0.3)
    ax.scatter(i_inter, j_inter, color='crimson', label='mir of interest')
    
    for nOrder, sNameMir in enumerate(name_inter):
        if np.isnan(i_inter[nOrder]) == True:
            continue
        if np.isnan(j_inter[nOrder]) == True:
            continue            
        ax.annotate(sNameMir, (i_inter[nOrder], j_inter[nOrder]))
    
    
    ax.set_xlim([2, 18])
    ax.set_ylim([2, 18])

    ax.set_title('{} vs. {} \nPearson={:.2f}, P={:.2f}'.format(i, j, fPearsonRho, fPearsonP))
    
    ax.set_xlabel('{} (log2 RPM)'.format(i))
    ax.set_ylabel('{} (log2 RPM)'.format(j))           
    ax.set_xticks(np.arange(2.5, 20, 2.5))
    ax.set_yticks(np.arange(2.5, 20, 2.5))
    
#plt.suptitle('Mature miRNA abundance', x=0.35, y=0.999)
plt.tight_layout()
#plt.show()

plt.savefig('/casa/jimkwon/PROJECT/ERH/results/200420_sRNAseq_RPM/scatter_293E_2.pdf', transparent=True)
plt.close()
```
