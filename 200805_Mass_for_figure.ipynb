To draw a plot of mass results

* 1st try: just DGCR8 IP (FLAG + STREP)
* 2nd try: control(293E) vs. DGCR8 IP



```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42

s1stDgcr8File = '/casa/jimkwon/PROJECT/ERH/results/200217_dgcr8mass/dgcr8mass_1st.txt'
s2ndControlFile = '/casa/jimkwon/PROJECT/ERH/results/200217_dgcr8mass/dgcr8mass_2nd_control.txt'
s2ndDgcr8File = '/casa/jimkwon/PROJECT/ERH/results/200217_dgcr8mass/dgcr8mass_2nd_dgcr8.txt'

fields = ['Protein name', 'Intensity', 'MS/MS Count']

df1stDgcr8 = pd.read_table(s1stDgcr8File, usecols=fields)
df2ndControl = pd.read_table(s2ndControlFile, usecols=fields)
df2ndDgcr8 = pd.read_table(s2ndDgcr8File, usecols=fields)

df2ndDgcr8
```

```python
def add_log2_intensity(df):
    df['Log2Intensity'] = df.apply(lambda row: np.log((row['Intensity'])) / np.log(2), axis=1)

for df in [df1stDgcr8, df2ndControl, df2ndDgcr8]:
    add_log2_intensity(df)

df2ndDgcr8
    
    
```

```python
df1stDgcr8 = df1stDgcr8.rename(columns={'Log2Intensity':'Log2Intensity_1stDgcr8'})
df2ndControl = df2ndControl.rename(columns={'Log2Intensity':'Log2Intensity_2ndControl'})
df2ndDgcr8 = df2ndDgcr8.rename(columns={'Log2Intensity':'Log2Intensity_2ndDgcr8'})

df2ndDgcr8
```

```python
dfAll = pd.merge(df1stDgcr8, df2ndDgcr8, how='outer', left_on='Protein name', right_on='Protein name')
dfAll = pd.merge(dfAll, df2ndControl, how='outer', left_on='Protein name', right_on='Protein name')
dfAll
```

```python
dfAll = dfAll.fillna(20)
dfAll
```

```python
dfAllLog = dfAll.loc[:,['Protein name', 'Log2Intensity_1stDgcr8', 'Log2Intensity_2ndDgcr8', 'Log2Intensity_2ndControl']]

dfAllLog
```

```python
l1stDgcr8 = dfAllLog['Log2Intensity_1stDgcr8'].tolist()
l2ndDgcr8 = dfAllLog['Log2Intensity_2ndDgcr8'].tolist()
l2ndControl = dfAllLog['Log2Intensity_2ndControl'].tolist()
lNames = dfAllLog['Protein name'].tolist()

# only presented their labels below
l1stDgcr8_only = l1stDgcr8[:4]
l2ndDgcr8_only = l2ndDgcr8[:4]
l2ndControl_only = l2ndControl[:4]
lNames_only = lNames[:4]

print(lNames_only)

```

```python
fig = plt.figure(figsize=(5,5))
plt.scatter(l1stDgcr8, l2ndDgcr8, c='k', alpha=0.5)
plt.scatter(l1stDgcr8_only, l2ndDgcr8_only, c='crimson')
plt.xlabel('DGCR8 IP, 1st trial (Log2 Intensity)')
plt.ylabel('DGCR8 IP, 2nd trial (Log2 Intensity)')
plt.title('DGCR8 IP, 1st vs. 2nd trials')

for x,y, name in zip(l1stDgcr8_only, l2ndDgcr8_only, lNames_only):
    if name == 'sp|P00761|TRYP_PIG':
        name = 'Trypsin'
        
    plt.annotate(name, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center


plt.xlim([19,40])
plt.ylim([19,40])
#plt.show()
plt.savefig('/casa/jimkwon/PROJECT/ERH/results/200217_dgcr8mass/dgcr8_1_2_5_crimson.pdf')
plt.close()
```

```python
fig = plt.figure(figsize=(5,5))
plt.scatter(l2ndControl, l2ndDgcr8, c='k', alpha=0.5)
plt.scatter(l2ndControl_only,l2ndDgcr8_only, c='crimson')

plt.xlabel('Control IP, 2nd trial (Log2 Intensity)')
plt.ylabel('DGCR8 IP, 2nd trial (Log2 Intensity)')
plt.title('DGCR8 IP vs. Control IP')

for x,y, name in zip(l2ndControl_only,l2ndDgcr8_only, lNames_only):
    if name == 'sp|P00761|TRYP_PIG':
        name = 'Trypsin'
        
    plt.annotate(name, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.xlim([19,40])
plt.ylim([19,40])
#plt.show()
plt.savefig('/casa/jimkwon/PROJECT/ERH/results/200217_dgcr8mass/dgcr8_control_5_2nd_crimson.pdf')
plt.close()
```
